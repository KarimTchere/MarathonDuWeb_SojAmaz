```python
from scipy.stats import chi2_contingency
import pandas as pd

# Charger les données
data = pd.read_csv("../Data/var_quali_no_binaire1.csv", sep=";", encoding='latin-1')

# Effectuer le test du chi-deux pour chaque paire de variables
resultats_chi2 = []
for colonne1 in data.columns:
    for colonne2 in data.columns:
        if colonne1 != colonne2:  # Éviter de comparer une variable avec elle-même
            table_contingence = pd.crosstab(data[colonne1], data[colonne2])
            chi2, p_value, dof, expected = chi2_contingency(table_contingence)
            resultats_chi2.append({'Variable 1': colonne1, 'Variable 2': colonne2, 'Chi2': chi2, 'P-value': p_value})

# Convertir les résultats en DataFrame pour une visualisation facile
resultats_chi2_df = pd.DataFrame(resultats_chi2)

# Afficher les résultats triés par p-value croissante
resultats_chi2_df.sort_values(by='P-value', ascending=True, inplace=True)
resultats_chi2_df
```

```python
data = pd.read_csv('var_quali_no_binaire1.csv', sep=";", encoding='latin-1')
data
matrice_correlation = data.corr(method='pearson')

# Tracer le heatmap avec seaborn
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(matrice_correlation, cmap='coolwarm')
plt.title('Heatmap de la matrice de corrélation')
plt.show()
```

```python
data
```

```python
# Filtrer les valeurs supérieures à 0.5
valeurs_superieures = matrice_correlation[(matrice_correlation > 0.5) & (matrice_correlation < 0.99)]

# Identifier les paires de variables associées à ces valeurs avec leurs coefficients de corrélation
paires_variables = [(colonne, ligne, valeurs_superieures.loc[ligne, colonne]) for colonne in valeurs_superieures.columns for ligne in valeurs_superieures.index if valeurs_superieures.loc[ligne, colonne] > 0.5]

# Créer un DataFrame à partir de la liste paires_variables
df_paires_variables = pd.DataFrame(paires_variables, columns=['Variable1', 'Variable2', 'Coefficient'])

# Exporter le DataFrame au format CSV
df_paires_variables.to_csv('paires_variables.csv', index=False)

print("DataFrame exporté avec succès.")
```
