```python
import matplotlib.pyplot as plt

# Diagramme en secteurs d'une variable qualitative
data['Q_3.1.3_genre'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Répartition par genre de la population')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Récupération des données
counts = data['Q_3.3.2_temps_communautŽs'].value_counts()

# Création d'une figure avec une taille personnalisée
plt.figure(figsize=(10, 8))

# Création de l'histogramme avec des barres horizontales
plt.barh(counts.index, counts.values)
plt.xlabel('Nombre de réponses')
plt.ylabel('Catégorie')

# Ajout des pourcentages sur les barres
for index, value in enumerate(counts):
    plt.text(value, index, f'{value} ({value / len(data) * 100:.1f}%)', va='center')

plt.title('Répartition par âge de la population')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Récupération des données
counts = data['Q_3.1.4_ScholaritŽ'].value_counts()

# Création d'une figure avec une taille personnalisée et un fond coloré
plt.figure(figsize=(10, 8))
plt.gca().set_facecolor('#f5f5f5')  # Couleur de fond gris clair

# Couleurs personnalisées pour les barres et le texte
bar_colors = ['#4682B4', '#32CD32', '#FFA07A', '#87CEEB', '#FFD700']
text_color = 'black'

# Création de l'histogramme avec des barres horizontales et des couleurs personnalisées
plt.barh(counts.index, counts.values, color=bar_colors)
plt.xlabel('Nombre de réponses', color=text_color)
plt.ylabel('Catégorie', color=text_color)

# Ajout des pourcentages sur les barres avec une police plus grande et en gras
for index, value in enumerate(counts):
    plt.text(value, index, f'{value} ({value / len(data) * 100:.1f}%)', va='center', color=text_color, fontsize=12, fontweight='bold')

plt.title("Niveau d'éducation", color=text_color, fontsize=14, fontweight='bold')
plt.gca().spines['top'].set_visible(False)  # Suppression des lignes du cadre en haut
plt.gca().spines['right'].set_visible(False)  # Suppression des lignes du cadre à droite
plt.show()
```

```python
import matplotlib.pyplot as plt

# Récupération des données pour le premier graphique
counts_1 = data['Q_9.6.2_changements_15ans_acteurs_extŽrieurs'].value_counts()

# Création de la première subplot
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)  # 1 ligne, 2 colonnes, premier graphique
plt.barh(counts_1.index, counts_1.values, color='skyblue')
plt.xlabel('Nombre de réponses')
plt.ylabel('Catégorie')
plt.title('Changement 15 ans acteurs extérieurs')

# Ajout des pourcentages sur les barres du premier graphique
for index, value in enumerate(counts_1):
    plt.text(value, index, f'{value} ({value / len(data) * 100:.1f}%)', va='center')

# Récupération des données pour le deuxième graphique
counts_2 = data['Q_9.6.2_changements_15ans_pouvoirs_publics'].value_counts()

# Création de la deuxième subplot
plt.subplot(1, 2, 2)  # 1 ligne, 2 colonnes, deuxième graphique
plt.barh(counts_2.index, counts_2.values, color='lightgreen')
plt.xlabel('Nombre de réponses')
plt.ylabel('Catégorie')
plt.title('Changement 15 ans pouvoirs publics')

# Ajout des pourcentages sur les barres du deuxième graphique
for index, value in enumerate(counts_2):
    plt.text(value, index, f'{value} ({value / len(data) * 100:.1f}%)', va='center')

# Ajustement de l'espacement entre les subplots
plt.tight_layout()

# Affichage de la figure contenant les deux graphiques
plt.show()
```
